```python
#for the sckeleton of the project
from crewai import Agent, Task, Crew, Process
#for llm setup
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
import os
from google.colab import userdata


#configuring our Google API Key
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
g_api_key = os.getenv("GOOGLE_API_KEY")
```

```python
#Setup the LLM
from langchain_google_genai import ChatGoogleGenerativeAI


my_llm = ChatGoogleGenerativeAI(model="gemini-pro",
                                verbose=True,
                                temperature=0.5,
                                google_api_key=g_api_key)
```

```python
#agent-1
senior_software_engineer = Agent(
    role="Senior Software Engineer",
    goal="Write clean and efficent code based on the problem statement.",
    backstory=(
    """You are a senior software engineer with ten years of working experience in the information technology industry.
    You can convert any logic into computer code.
    Based on the problem statement, your work is to write a clean & correct end-to-end solution code
    in the programming language mentioned by the user."""),
    llm=my_llm,
    allow_delegation=False,
    verbose=True
)


#agent-2
senior_software_testing_engineer = Agent(
    role="Senior Software Testing Engineer",
    goal="Test the solution codes written by the senior_software_engineer.",
    backstory=(
    """You are a senior software testing engineer with ten years of working experience in the information technology industry.
    You are an expert in finding bugs in the computer code and fixing it.
    If there is a bug in the code then write the correct code otherwise pass the code to the next agent."""),
    llm=my_llm,
    allow_delegation=False,
    verbose=True
)


#agent-
coding_documentation_specialist = Agent(
    role="Senior Coding Documentation Specialist",
    goal="Write the documention of the code passed by the senior_software_testing_engineer.",
    backstory=(
    """You are an expert coding documentation specialist, working for years in the IT industry.
    Read the code and write a clear step by step documentation of the code.
    Include clear and detailed instructiuons about using the code."""),
    llm=my_llm,
    allow_delegation=False,
    verbose=True
)
```

```python
task1 = Task(
    description=(
        "1. Analyze the problem statement provided by the user and identify the key requirements for the code."
        "2. Write a clean and efficient end-to-end solution code in the programming language mentioned by the user."
        "3. Ensure that the code is syntactically correct and follows best practices."
        "4. If necessary, optimize the code for better performance."
        "5. Pass the final code to the senior_software_testing_engineer for testing."
    ),
    expected_output=(
        "A complete, clean, and correct code solution that meets the problem statement's requirements,"
        "written in the specified programming language."
    ),
    agent=senior_software_engineer,
    human_input=True,
)


task2 = Task(
    description=(
        "1. Review the code provided by the senior_software_engineer to identify any bugs or issues."
        "2. If a bug or issue is found, fix the code and ensure it works correctly."
        "3. Run various tests to validate the functionality and performance of the code."
        "4. If the code passes all tests and is bug-free, pass it to the coding_documentation_specialist."
        "5. Provide a summary of the testing process and any changes made to the code."
    ),
    expected_output=(
        "A bug-free and fully tested code solution that is ready for documentation."
    ),
    agent=senior_software_testing_engineer,
    human_input=False,
)


task3 = Task(
    description=(
        "1. Review the code passed by the senior_software_testing_engineer."
        "2. Write a clear and detailed step-by-step documentation of the code."
        "3. Include instructions on how to use the code, along with examples if necessary."
        "4. Ensure that the documentation is easy to understand and follow."
        "5. Finalize the documentation and prepare it for delivery to the user."
        "6.Return the code and the documentation."
    ),
    expected_output=(
        "The code and a comprehensive and well-structured documentation that explains the code, its usage,"
        "and includes examples if needed."
    ),
    agent=coding_documentation_specialist,
    human_input=False,
)
```

```python
crew = Crew(
  agents=[senior_software_engineer, senior_software_testing_engineer, coding_documentation_specialist],
  tasks=[task1, task2, task3],
  verbose=True
)
```

```python
topic = input("Enter your probelm statement here: ")
result = crew.kickoff(inputs={"topic": topic})
```
